```python
from ast import literal_eval

tmdb_1['genres'] = tmdb_1['genres'].apply(literal_eval)
```

```python
tmdb_1.head()
```

```python
tmdb_1['genres'] = tmdb_1['genres'].apply(lambda x:[y['name'] for y in x])
tmdb_1['genres'].head()
```

```python
tmdb_1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

tmdb_1['genres_literal'] = tmdb_1['genres'].apply(lambda x:('').join(x))
count_vect = CountVectorizer(min_df = 0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(tmdb_1['genres_literal'])
```

```python
# 투표 수 가중평균
m = tmdb_1['vote_count'].quantile(0.6)
c = tmdb_1['vote_average'].mean()

def weight_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ((v/(v+m)) * R) + ((m/(m+v)) * c)

tmdb_1['w_vote'] = tmdb_1.apply(weight_vote_average, axis = 1)
```

```python
tmdb_1[['title','w_vote', 'vote_average','vote_count']].sort_values('w_vote', ascending = False)[:10]
```

```python
# 코사인 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
```

```python
print(genre_sim[:1])
```

```python
genre_sim_sorted_index = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_index[:2])
```

```python
title_movie = tmdb_1['title'] 
print(title_movie.index)
print(title_movie.index.values)
```

```python
def find_sim_movie(tmdb_1, sorted_index, title_name, top_n = 5):
    title_movie = tmdb_1[tmdb_1['title']==title_name]
    title_index = title_movie.index.values
    similar_ind = sorted_index[title_index, : (top_n)]
    print(similar_ind)
    similar_ind = similar_ind.reshape(-1)
    
    return tmdb_1.iloc[similar_ind]
    
```

```python
similar_movies = find_sim_movie(tmdb_1, genre_sim_sorted_index, 'Iron Man')
similar_movies[['title', 'w_vote']]
```
